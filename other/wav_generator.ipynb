```python
# from https://stackoverflow.com/questions/33213408/python-convert-an-array-to-wav

import scipy.io.wavfile as wavf
import numpy as np
import matplotlib.pyplot as plt
```

```python
SAMPLE_RATE = 48000
RESOLUTION = 2048

t = np.linspace(0, 2*np.pi, RESOLUTION, dtype=np.float32)
# samples = t * 0
# out_file = f"../resources/wavetables/32_bit/{SAMPLE_RATE}/zeros.wav"
# samples = np.sin(t)
# out_file = f"../resources/wavetables/32_bit/{SAMPLE_RATE}/sine.wav"
# samples = np.sign(t-np.pi)
# out_file = f"../resources/wavetables/32_bit/{SAMPLE_RATE}/square.wav"

t = np.linspace(0, 1, RESOLUTION, dtype=np.float32)
samples = 2 * (t - np.floor(1/2 + t))
out_file = f"../resources/wavetables/32_bit/{SAMPLE_RATE}/sawtooth.wav"
# samples = 4 * np.abs(t - np.floor(1/2 + t)) - 1
# out_file = f"../resources/wavetables/32_bit/{SAMPLE_RATE}/triangle.wav"

wavf.write(out_file, SAMPLE_RATE, samples)
```
